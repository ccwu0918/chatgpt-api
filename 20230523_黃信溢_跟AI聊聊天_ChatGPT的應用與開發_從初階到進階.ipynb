```python
# 製作可以查詢字典的互動程式
# 程式顯示「請輸入要查詢的國字:」
# 完成顯示查詢結果

import requests
w = input("請輸入要查詢的國字:")
url = 'https://www.moedict.tw/uni/' + w
r = requests.get(url)
datas = r.json()
print('國字:', datas['title'])
print('部首:', datas['radical'])
print('筆劃:', datas['stroke_count'])
print()
for i in range(len(datas['heteronyms'])):
    print('注音:', datas['heteronyms'][i]['bopomofo'])
    print('拼音:', datas['heteronyms'][i]['pinyin'])
    print()
```

```python
# 製作可以查詢字典的互動程式-用自訂函式的方法
# checkword(w)，回傳值即是查詢結果
# 程式顯示「請輸入要查詢的國字:」
# 完成顯示查詢結果

import requests

def checkword(w):
    url = 'https://www.moedict.tw/uni/' + w
    r = requests.get(url)
    datas = r.json()
    msg = '國字:' + datas['title'] + '\n'
    msg += '部首:' + datas['radical'] + '\n'
    msg += '筆劃:' + str(datas['stroke_count']) + '\n'
    for i in range(len(datas['heteronyms'])):
        msg += '注音:' + datas['heteronyms'][i]['bopomofo'] + '\n'
        msg += '拼音:' + datas['heteronyms'][i]['pinyin'] + '\n'
    return msg

w = input("請輸入要查詢的國字:")
print(checkword(w))
```

```python
# 製作可以查詢字典的WebAPP

import requests
import gradio as gr

def checkword(w):
    url = 'https://www.moedict.tw/uni/' + w
    r = requests.get(url)
    datas = r.json()
    msg = '國字:' + datas['title'] + '\n'
    msg += '部首:' + datas['radical'] + '\n'
    msg += '筆劃:' + str(datas['stroke_count']) + '\n'
    for i in range(len(datas['heteronyms'])):
        msg += '注音:' + datas['heteronyms'][i]['bopomofo'] + '\n'
        msg += '拼音:' + datas['heteronyms'][i]['pinyin'] + '\n'
    return msg

gui = gr.Interface(
    fn = checkword,
    inputs = 'text',
    outputs = 'text'
)
gui.launch(share=True)
```
